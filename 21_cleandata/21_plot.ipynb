```python
import sqlite3
import matplotlib.pyplot as plt

# Connect to the new SQLite database
new_conn = sqlite3.connect('Normal_Zustand.db')
new_cursor = new_conn.cursor()

# Retrieve the timestamp and still data from the table
new_cursor.execute("SELECT timestamp, zustand FROM mockdata")
data = new_cursor.fetchall()

# Close the connection
new_cursor.close()
new_conn.close()

# Separate the data into two lists
timestamps, stills = zip(*data)

# Plot the data
plt.figure(figsize=(15, 7))
plt.plot(timestamps, stills, marker='o', linestyle='-', color='b')
plt.xlabel('Timestamp')
plt.ylabel('Zustand')
plt.title('Relation between Timestamp and Zustand')
plt.grid(True)
plt.show()
```

```python
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
new_conn = sqlite3.connect('Normal_GX.db')
new_cursor = new_conn.cursor()

# Retrieve the timestamp and still data from the table
new_cursor.execute("SELECT n, zustand FROM mockdata")
data = new_cursor.fetchall()

# Close the database connection
new_conn.close()

# Extract n and zustand values
n_values = [row[0] for row in data]
zustand_values = [row[1] for row in data]

# Plot the scatter plot
plt.figure(figsize=(20, 6))
plt.scatter(n_values, zustand_values, color='r', marker='o', s=0.8)
plt.title('Zustand Values over Index (n)')
plt.xlabel('Index (n)')
plt.ylabel('Zustand')
plt.yticks([0, 1])
plt.grid(True)
plt.show()
```

```python
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
new_conn = sqlite3.connect('Anomalie_GX.db')
new_cursor = new_conn.cursor()

# Retrieve the zustand data from the table
new_cursor.execute("SELECT zustand FROM mockdata")
data = new_cursor.fetchall()

# Close the database connection
new_conn.close()

# Extract zustand values
zustand_values = [row[0] for row in data]

# Plot the scatter plot over the default index
plt.figure(figsize=(20, 6))
plt.scatter(range(len(zustand_values)), zustand_values, color='r', marker='o', s=0.8)
plt.title('Zustand Values over Default Index')
plt.xlabel('Default Index')
plt.ylabel('Zustand')
plt.yticks([0, 1])
plt.grid(True)
plt.show()
```

```python
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
new_conn = sqlite3.connect('Training_Zustand.db')
new_cursor = new_conn.cursor()

# Retrieve zustand and label data from the table
new_cursor.execute("SELECT zustand, label FROM mockdata")
data = new_cursor.fetchall()

# Close the database connection
new_conn.close()

# Extract zustand and label values
zustand_values = [row[0] for row in data]
labels = [row[1] for row in data]

# Determine colors based on label
colors = ['b' if label == 'normal' else 'r' for label in labels]

# Plot the scatter plot
plt.figure(figsize=(20, 6))
plt.scatter(range(len(zustand_values)), zustand_values, color=colors, marker='o', s=0.8)
plt.title('Zustand Values with Label Color Coding')
plt.xlabel('Default Index')
plt.ylabel('Zustand')
plt.yticks([0, 1])
plt.grid(True)
plt.show()
```
